```python
import time
from google.colab import drive
from google.colab import runtime

print("Experiment complete. Preparing to pull the plug...")

# 1. Flush the buffer (CRITICAL)
# This forces Google Drive to finish writing your CSVs before the machine dies
drive.flush_and_unmount()
print("Drive unmounted safely.")

# 2. Wait 10 seconds just to be absolutely sure network I/O is done
time.sleep(10)

# 3. Pull the plug
print("Terminating session and releasing A100 GPU. Goodnight!")
runtime.unassign()
```
