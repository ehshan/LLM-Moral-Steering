# Moral Machine Dataset Analysis

This notebook downloads the Moral Machine dataset directly from the Open Science Framework (OSF), unzips it, and loads it into a pandas DataFrame for preliminary analysis.

### Step 1: Download the Dataset

We will use the `wget` command to download the compressed data file from the URL. This might take a few minutes as the file is quite large (~1.5 GB).

```python
# The URL provided for the data file
url = 'https://osf.io/3hvt2/files/osfstorage/5b54f679c86a8c0010444782'

# Use wget to download the file. The -O flag lets us specify the output filename.
!wget -O MoralMachineData.zip {url}
```

### Step 2: Unzip the File

Now that we have the `.zip` file, we'll use the `unzip` command to extract its contents.

```python
!unzip MoralMachineData.zip
```

### Step 3: Check the Extracted Files

Let's list the files in our current directory to see what was extracted. This will help us find the name of the main `.csv` data file.

```python
# List files with human-readable sizes
!ls -lh
```

### Step 4: Load the Data into a pandas DataFrame

From the output above, we can see the main data file. We'll now load this CSV file into a pandas DataFrame.

**Note:** Loading this large file into memory can take a moment.

```python
import pandas as pd

# From the previous step, we can identify the correct filename.
# Make sure this filename matches the output from the `ls -lh` command.
file_name = 'MoralMachineData.csv'

print(f"Loading {file_name}...")
df = pd.read_csv(file_name)
print("Data loaded successfully!")
```
